```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('Online_Retail1.csv',encoding='latin1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(axis=0,subset=['InvoiceNo'],inplace=True) # drop NA if this col is empty
```

```python
df['Description']=df['Description'].str.strip()    # remove spaces
```

```python
df['InvoiceNo']=df['InvoiceNo'].astype('str')    # convert to string
```

```python
df['InvoiceNo'].nunique()
```

```python
#list(df['InvoiceNo'].unique())
```

```python
df=df[~df['InvoiceNo'].str.contains('C')] ## remove those invoice ID with C
```

```python
df.info()
```

```python
df['Country'].unique()
```

```python
#df.groupby(by='Country',axis=0)['InvoiceNo'].count()
df['Country'].value_counts()
```

```python
df['Country'].nunique()
```

```python
basket=df[df['Country']=='France'].groupby(by=['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
basket
```

```python
#list(basket.columns)
```

```python
basket.drop('POSTAGE',inplace=True,axis=1) # THIS IS ONLY FOR EXAMPLE DISPLAY. THIS IS SAY WE NEED TO CHECK MANUALLY WHICH ARE NOT NECESSARY
```

```python
## we are not interested in quantity, we only want to know has or not

def encode_units(x):
    if x<1:
        return 0
    else:
        return 1
```

```python
basket_sets=basket.applymap(encode_units)
```

```python
basket_sets
```

```python
frequent_itemsets=apriori(basket_sets,min_support=0.07, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets, metric='lift',min_threshold=1)
```

```python
rules
```

```python
df=pd.read_csv('Recommend.csv',names=['user_id','movie_id','rating','ts'])
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
n_user=df.user_id.nunique()
```

```python
n_movie=df.movie_id.nunique()
```

```python
train_data,test_data=train_test_split(df,test_size=0.3)
```

```python
train_data.info()
```

```python
train_data_matrix = np.zeros((n_user,n_movie))
```

```python
train_data_matrix
```

```python
for line in train_data.itertuples():
    #print(line)
    train_data_matrix[line[1]-1,line[2]-1]=line[3]
```

```python
train_data_matrix=pd.DataFrame(train_data_matrix)
```

```python
train_data_matrix
```

```python
test_data_matrix = np.zeros((n_user,n_movie))

for line in test_data.itertuples():
    #print(line)
    test_data_matrix[line[1]-1,line[2]-1]=line[3]

test_data_matrix
```

```python
from sklearn.metrics import pairwise_distances
```

```python
user_similarity=pairwise_distances(train_data_matrix,metric='cosine')
movie_similarity=pairwise_distances(train_data_matrix.T,metric='cosine')
```

```python
mean_user_ratings = train_data_matrix.mean(axis=1)[:,np.newaxis]
#mean_user_ratings 
```

```python
rating_diff= (train_data_matrix - mean_user_ratings)
```

```python
user_pred = mean_user_ratings + user_similarity.dot(rating_diff) / np.array([np.abs(user_similarity).sum(axis=1)]).T
```

```python
user_pred.shape
```

```python
movie_pred=train_data_matrix.dot(movie_similarity) / np.array([np.abs(movie_similarity).sum(axis=1)])
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(pred, test):
    pred=pred[test.nonzero()].flatten()  #. test.nonzero() gives us indice
    test=test[test.nonzero()].flatten()
    return sqrt(mean_squared_error(pred,test))
```

```python
print('RMSE for 30000 records using UBCF is: ', rmse(user_pred,test_data_matrix))
```

```python
print('RMSE for 30000 records using CBF is: ', rmse(movie_pred,test_data_matrix))
```
