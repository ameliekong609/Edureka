```python
class MyDt():
    def __init__(self,d,m,y):
        print('This is MyDt class instance created using dmy format')
        self.date=d
        self.month=m
        self.year=y
    @classmethod
    def mdy(self, m,d,y):
        self.month=m
        self.date=d
        self.year=y     
        return self(self.date, self.month, self.year)   
    def today(self):
        print('This is base class fun')
    def __del__(self):
        print("xxx")
```

```python
#dt1=MyDt(19,11,2021)
dt2=MyDt.mdy('Oct',4,2022)
dt2
```
