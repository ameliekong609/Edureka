```python
# define missing values
missing_values=["n/a","na","--"]
data=pd.read_csv("Sample_real_estate_data.csv",na_values=missing_values)
print(data['NUM_BEDROOMS'].isnull())
```

```python
# check another type of missing var - numerical within text
print(data["OWN_OCCUPIED"].isnull())
# actually the NO.3 is a numerical number, and it should be detected
```

```python

#another type of missing values e.g. suden number in a text column
count=0
for row in data["OWN_OCCUPIED"]:
    try :
        int(row)
        data.loc[count,"OWN_OCCUPIED"]=np.nan
    except ValueError:
        pass
    count+=1
```

```python
## how many missing values by var
print(data.isnull().sum())
```

```python
#pip install -U scikit-learn scipy matplotlib
```

```python
## impute missing data with mean value
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer=imputer.fit(data[["SQ_FT"]])
data["SQ_FT"]= imputer.transform(data[["SQ_FT"]])
```

```python
## impute data of "numbers of bedrooms" with mediam values
print(data["NUM_BEDROOMS"])
median=data["NUM_BEDROOMS"].median()
print(median)
data["NUM_BEDROOMS"]=data["NUM_BEDROOMS"].fillna(median)
print(data["NUM_BEDROOMS"])
```
